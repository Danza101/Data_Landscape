```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Initial_Breakdown_Plan.xlsx - Dataset.csv')
```

```python
# Getting the amount of items purchased in each area
london = df[df['Location'] == 'London']
belfast = df[df['Location'] == 'Belfast']
reading = df[df['Location'] == 'Reading']
birmingham = df[df['Location'] == 'Birmingham']
leeds = df[df['Location'] == 'Leeds']
bristol = df[df['Location'] == 'Bristol']
manchester = df[df['Location'] == 'Manchester']
watford = df[df['Location'] == 'Watford']
edinburgh = df[df['Location'] == 'Edinburgh']
glasgow  = df[df['Location'] == 'Glasgow']
cardiff  = df[df['Location'] == 'Cardiff']
```

```python
london_daily_totals = london.groupby('Day')['No of same item bought'].sum().reset_index()
belfast_daily_totals = belfast.groupby('Day')['No of same item bought'].sum().reset_index()
reading_daily_totals = reading.groupby('Day')['No of same item bought'].sum().reset_index()
birmingham_daily_totals = birmingham.groupby('Day')['No of same item bought'].sum().reset_index()
leeds_daily_totals = leeds.groupby('Day')['No of same item bought'].sum().reset_index()
bristol_daily_totals = bristol.groupby('Day')['No of same item bought'].sum().reset_index()
manchester_daily_totals = manchester.groupby('Day')['No of same item bought'].sum().reset_index()
watford_daily_totals = watford.groupby('Day')['No of same item bought'].sum().reset_index()
edinburgh_daily_totals = edinburgh.groupby('Day')['No of same item bought'].sum().reset_index()
glasgow_daily_totals = glasgow.groupby('Day')['No of same item bought'].sum().reset_index()
cardiff_daily_totals = cardiff.groupby('Day')['No of same item bought'].sum().reset_index()
```

```python
def items_brought_per_week(location):
    
    # This line groups the dataframe by 'Week', as you've stated
    weeks = location.groupby('Week')

    # Now, to get the number of items bought per week, you'll need to sum the 'No of same item bought' column
    weekly_items_bought = weeks['No of same item bought'].sum()

    # To view the result as a DataFrame
    weekly_items_bought_df = weekly_items_bought.reset_index()

    # Display the DataFrame
    return(weekly_items_bought_df)
```

```python
london_weekly_purchase = items_brought_per_week(london)
belfast_weekly_purchase = items_brought_per_week(belfast)
reading_weekly_purchase = items_brought_per_week(reading)
birmingham_weekly_purchase = items_brought_per_week(birmingham)
leeds_weekly_purchase = items_brought_per_week(leeds)
bristol_weekly_purchase = items_brought_per_week(bristol)
manchester_weekly_purchase = items_brought_per_week(manchester)
watford_weekly_purchase = items_brought_per_week(watford)
belfast_weekly_purchase = items_brought_per_week(belfast)
edinburgh_weekly_purchase = items_brought_per_week(edinburgh)
glasgow_weekly_purchase = items_brought_per_week(glasgow)
```

```python
total_avg_weekly_purchase = []

for i in range(52):  # This assumes week numbers start at 0 and go up to 51
    total_avg_weekly_purchase.append(
        (
            belfast_weekly_purchase.iloc[i] +
            reading_weekly_purchase.iloc[i] +
            birmingham_weekly_purchase.iloc[i] +
            leeds_weekly_purchase.iloc[i] +
            bristol_weekly_purchase.iloc[i] +
            manchester_weekly_purchase.iloc[i] +
            watford_weekly_purchase.iloc[i] +
            edinburgh_weekly_purchase.iloc[i] +
            glasgow_weekly_purchase.iloc[i]
        ) / 9  # Divided by 9, since Belfast is mentioned only once now
    )
```

```python
sum_of_weekly_purchase = belfast_weekly_purchase + reading_weekly_purchase + birmingham_weekly_purchase + leeds_weekly_purchase + bristol_weekly_purchase + manchester_weekly_purchase + watford_weekly_purchase + edinburgh_weekly_purchase + glasgow_weekly_purchase
```
